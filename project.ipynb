```python
print('Gender ',df['Gender'].unique())
print('Customer Type ',df['Customer Type'].unique())
print('Type of Travel ',df['Type of Travel'].unique())
print('Class ',df['Class'].unique())
print('satisfaction ',df['satisfaction'].unique())
```

```python
def gender(x):
    if x == 'Male':
        return 0
    elif x == 'Female':
        return 1

def Customer_Type(x):
    if x == 'Loyal Customer':
        return 0
    elif x == 'disloyal Customer':
        return 1

def Type_of_Travel(x):
    if x == 'Personal Travel':
        return 0
    elif x == 'Business travel':
        return 1

def Class(x):
    if x == 'Eco Plus':
        return 2
    elif x == 'Business':
        return 1
    elif x == 'Eco':
        return 0

def satisfaction(x):
    if x == 'neutral or dissatisfied':
        return 0
    elif x == 'satisfied':
        return 1
```

```python
df['Gender'] = df['Gender'].map(gender)
df['Customer Type'] = df['Customer Type'].map(Customer_Type)
df['Type of Travel'] = df['Type of Travel'].map(Type_of_Travel)
df['Class'] = df['Class'].map(Class)
df['satisfaction'] = df['satisfaction'].map(satisfaction)
```

```python
plt.figure(figsize=(15,17))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
```

```python
df.drop(['Departure Delay in Minutes'],axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop('satisfaction', axis=1)
y = df['satisfaction']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, stratify=y, random_state=1)
```

```python
print('x_train shape: ', x_train.shape)
print('x_test shape: ', x_test.shape)
print('y_train shape: ', y_train.shape)
print('y_test shape: ', y_test.shape)
```

```python
std = StandardScaler()

x_train_scaled = std.fit_transform(x_train)
x_test_scaled = std.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
import pickle
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
def classification_report(model, x_test, y_test, model_name):

    print(f'-------------------{model_name}-------------------')
    y_pred = model.predict(x_test)

    print('Accuracy: ', accuracy_score(y_test, y_pred))
    print('Recall: ', recall_score(y_test, y_pred))
    print('Precision: ', precision_score(y_test, y_pred))
    print('f1-score: ', f1_score(y_test, y_pred))

    cm = confusion_matrix(y_test,y_pred)

    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')

    tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()

    print('The accuracy of prediction for class 0 is: ', tp/(tp+fp))
    print('The accuracy of prediction for class 1 is: ', tn/(tn+fn))

    with open(f'{model_name}.pickle', 'wb') as f:
        pickle.dump(model,f)
```

```python
XGB_params = {
    'n_estimators':[10,30,50,70,90,110,130,150,170],
    'max_depth':[1,2,3,4,5,6],
    'learning_rate':[0.1,0.01,0.001]
}
```

```python
XGB_opt = GridSearchCV(XGBClassifier(use_label_encoder=False), XGB_params, scoring='accuracy', cv=5, return_train_score=False)
XGB_opt.fit(x_train_scaled, y_train)
```

```python
print(XGB_opt.best_params_)
print(XGB_opt.best_score_)
```

```python
classification_report(XGB_opt.best_estimator_, x_test_scaled, y_test, 'XGBClassifier')
```

```python
random_params = {
    'n_estimators':[10,30,50,70,90,110,130,150,170],
    'criterion':['gini', 'entropy', 'log_loss'],
    'max_features':['sqrt', 'log2', None]
}
```

```python
rf_opt = GridSearchCV(RandomForestClassifier(), random_params, scoring='accuracy', cv=5, return_train_score=False)
rf_opt.fit(x_train_scaled, y_train)
```

```python
print(rf_opt.best_params_)
print(rf_opt.best_score_)
```

```python
classification_report(rf_opt.best_estimator_, x_test_scaled, y_test, 'Random forest')
```

```python
knn_params = {
    'n_neighbors':[10,30,50,70,90,110,130,150,170],
    'metric':['euclidean', 'manhattan', 'chebyshev','minkowski', 'wminkowski', 'seuclidean', 'mahalanobis'],
}
```

```python
knn_opt = GridSearchCV(KNeighborsClassifier(), knn_params, scoring='accuracy', cv=5, return_train_score=False)
knn_opt.fit(x_train_scaled, y_train)
```

```python
print(knn_opt.best_params_)
print(knn_opt.best_score_)
```

```python
classification_report(knn_opt.best_estimator_, x_test_scaled, y_test, 'KNN')
```
